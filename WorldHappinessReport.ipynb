```python
# - Import the essential libraries required for this analysis using the following commands:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the downloaded dataset into a pandas DataFrame using the following command:
df = pd.read_csv('world-happiness-report-2019.csv')
```

```python
# Get an initial feel of the dataset by looking at the first few rows, columns, and summary statistics:
df.head()
df.info()
df.describe()
```

```python
# Check for missing values and handle them accordingly. You can drop or fill the missing values based on the dataset:
df.isna().sum()
```

```python
# Start by plotting the happiness score distribution:
sns.histplot(df['Score'], kde=True)
plt.title('Happiness Score Distribution')
plt.show()
```

```python
# Calculate the correlation matrix and visualize it using a heatmap:

corr_matrix = df.drop("Country or region",axis=1).corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Analyze the relationship between happiness score and key factors
# Create scatter plots to visualize the relationship between the identified key factors and happiness score:

sns.scatterplot(x='GDP per capita', y='Score', data=df)
plt.title('Happiness Score vs GDP per capita')
plt.show()
```

```python
# Analyze happiness levels by region
# Create a boxplot to visualize the happiness score distribution within each region:
sns.boxplot(x='Score', y='Country or region', data=df, orient='h')
plt.title('Happiness Score Distribution by Region')
plt.show()
```

```python
# Identify the top factors contributing to happiness
# - Analyze the top factors contributing to happiness by looking at their respective average values for the highest-scoring countries:

top_countries = df.drop("Country or region",axis=1).nlargest(10, 'Score')
top_countries.mean().sort_values(ascending=False)
```
